```python
import sys
import os

# Add project root to Python path
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path:
    sys.path.append(project_root)

# Now you can import from src
from src.utils.db_inventory_manager import DBInventoryManager
from src.utils.db_transactions_manager import DBTransactionsManager
import pandas as pd

print("Path setup complete!")
```

```python
# Initialize managers
inventory_manager = DBInventoryManager()
transactions_manager = DBTransactionsManager()

# Get inventory data
inventory_data = inventory_manager.get_inventory()
columns = ['item_id', 'item_name', 'product_type', 'quantity', 'price', 'unit_cost', 'medicine_type', 'expiration_date']
inventory_df = pd.DataFrame(inventory_data, columns=columns)

# Get transaction data
transaction_data = transactions_manager.get_transactions()
trans_columns = ['transaction_id', 'item_name', 'quantity_bought', 'unit_price', 'total_price', 'date_processed', 'transaction_item_id']
transaction_df = pd.DataFrame(transaction_data, columns=trans_columns)

print(f"Inventory: {len(inventory_df)} items")
print(f"Transactions: {len(transaction_df)} records")
```

```python
# Data Analysis Examples

# Add calculated columns
inventory_df['total_value'] = inventory_df['quantity'] * inventory_df['price']
inventory_df['profit_margin'] = (inventory_df['price'] - inventory_df['unit_cost']) / inventory_df['unit_cost']

# Convert date column
transaction_df['date_processed'] = pd.to_datetime(transaction_df['date_processed'])

# Low stock analysis
low_stock = inventory_df[inventory_df['quantity'] < 10]
print(f"Low stock items: {len(low_stock)}")
print(low_stock[['item_name', 'quantity']].head())
```

```python
# Sales analysis
daily_sales = transaction_df.groupby(transaction_df['date_processed'].dt.date)['total_price'].sum()

print("Daily Sales:")
print(daily_sales.tail(10))

# Top selling products
top_products = transaction_df.groupby('item_name')['quantity_bought'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Products:")
print(top_products)
```
